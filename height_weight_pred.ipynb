```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.info()
```

```python
df.head(2)
```

```python
X = df.Height.values  # input data
```

```python
print(X[:5])
```

```python
type(X)
```

```python
y_true = df.Weight.values  # labels , output, answers
```

```python
print(y_true[:5])
```

```python
# line equation 
def line(x, W=0, b=0):
    return x* W + b
```

```python
# cost function
def mean_squared_error(y_true, y_pred):
    s = (y_true - y_pred) **2
    return np.sqrt(s.mean())
```

```python
plt.figure(figsize=(10,5))
ax1 = plt.subplot(121)
df.plot(kind = 'scatter',x='Height', y='Weight', title='Height vs Weight')
```

```python
bs = np.random.choice([-300,-150,-100,-50,0,50,100,150,300],10)
```

```python
bs
```

```python
plt.figure(figsize=(10,5))
ax1 = plt.subplot(121)
df.plot(kind = 'scatter',x='Height', y='Weight', title='Height vs Weight', ax = ax1)

mse = []
for b in bs:
    y_pred = line(X, W=2, b = b)
    msev = mean_squared_error(y_true, y_pred)
    mse.append(msev)
    plt.plot(X,y_pred)
    plt.show
```

```python
ax2 = plt.subplot(122)
plt.plot(bs, mse,'o-')
plt.title('Cost Functions For Each Line')
plt.xlabel('b')
```
